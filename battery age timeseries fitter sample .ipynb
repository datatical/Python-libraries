```python
import numpy as np
import pandas as pd
import seaborn as sns
from fitter import Fitter, get_common_distributions
# Now you can use get_common_distributions
distributions = get_common_distributions()
print(distributions)
```

```python
dataset = pd.read_csv("age_timeseries.csv")
dataset.head()
```

```python
dataset.info
```

```python
sns.set_style('dark')
sns.set_context("paper", font_scale = 2)
sns.displot(data=dataset, x="Constant Voltage Charge Energy (Wh)", kind="hist", bins = 100, aspect = 1.5)
```

```python
# Check the column names in your dataset DataFrame
print(dataset.columns)

# Specify the correct column name for the x parameter
sns.displot(data=dataset, x="Total Charge Time (s)", kind="hist", bins=100, aspect=1.5)
```

```python
 height = dataset["Cycle Start Time (s)"].values 
```

```python
f = Fitter(height,
           distributions=['gamma',
                          'lognorm',
                          "beta",
                          "burr",
                          "norm"])
f.fit()
f.summary()
```

```python
f.get_best(method = 'sumsquare_error')
```

```python
#identifying the parameters
f.fitted_param["beta"]
```

```python
#Fitting Distributions on a Randomly Drawn Data
#Common distributions --the Fitter library also has a get_common_distribution( ) method
get_common_distributions()
```

```python
#Generating data from the normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
data = np.random.normal(mu, sigma, 10000)
data
```

```python
# the sumsquared_error criteria the best-fitted distribution is the normal distribution
f = Fitter(data,
           distributions= get_common_distributions())
f.fit()
f.summary()
```

```python
#Get best-fitted distribution and parameters
f.get_best(method = 'sumsquare_error')
```
